```python
import PyPDF2
```

```python
#Create a python file object variable
#opening method will be rb (reading binary)
pdffileobj=open('FacturaEjemploStatAIEnglish.pdf','rb')
 
#Create reader variable that will read the pdffileobj
pdfreader=PyPDF2.PdfReader(pdffileobj)
 
#This will store the page in a page object. The 0 means that we're reading only one page
pageobj=pdfreader.pages[0]

#Create text variable which will store all text datafrom pdf file
text=pageobj.extract_text()
```

```python
#Print the pdf text
print(text)
```
